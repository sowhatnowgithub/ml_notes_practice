# Histogram 


```python
from collections import Counter
```

```python
grades = []
from random import randint
for i in range(15):
    rand = randint(0,100)
    grades.append(rand)
len(grades)
```

```python
histogram = Counter(min(grade // 10 * 10 ,90 ) for grade in grades)
histogram
```

```python
plt.bar([i+5 for i in histogram.keys()], [histogram[i] for i in histogram])
plt.xticks([10*i for i in range(11)])
plt.axis([0,105,0,5])
```

```python
plt.bar([i+5 for i in histogram.keys()], histogram.values(), 10, edgecolor=(0,0,0))
plt.xlabel('Grades')
plt.ylabel('#students')
plt.axis([-5, 105, 0, 5])
plt.xticks( [ 10*i for i in range(11)])
plt.show()
```

```python
mentions = [200, 505]
years = [2016, 2017]
plt.bar( years, mentions,0.8,edgecolor = (0,0,0))
plt.xticks(years)
plt.axis([2015.5,2017.5,0, 550])
plt.show()
```

# Line charts

```python
variance = [1]
for i in range(1,10):
    variance.append(2**i)
print(variance)
bias_squared = []
for i in variance[::-1]:
    bias_squared.append(i)

for i in reversed(variance):
    bias_squared.append(i)

bias_squared = [i for i in reversed(variance)]
print(bias_squared)
```

```python
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for _,i in enumerate(variance)]
print(xs)
xs = [i for i,_ in enumerate(variance)]
print(xs)
```

```python
plt.plot(xs, variance, 'b.',label='Variance')
plt.plot(xs, bias_squared, color='green',marker='.',label='Bias')
plt.plot(xs, total_error, 'r-',label='total_error')
plt.xticks([])
plt.xlabel('Model Compelxity')
plt.title('The Bias-Variance Tradeoff')
plt.legend()
plt.show()
```

```python
from random import randint as rand
minutes = []
friends = []
for i in range(10):
    friends.append(rand(50,100))
    minutes.append(rand(100,200))
labels = ['a','b','c','d','e','f','g','h','j','k']
len(labels)
```

```python
plt.scatter(friends, minutes)
plt.xlabel('Friend id')
plt.ylabel('Minutes')
plt.title("Friend And time spent")
```

```python
plt.scatter(friends,minutes)
for label,friend_id, minute in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy = (friend_id,minute),
                xytext = (5,-5),
                textcoords='offset points')
    
plt.show()
```
