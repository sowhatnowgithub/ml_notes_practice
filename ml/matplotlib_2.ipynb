```python
from collections import Counter
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
histogram = Counter(min(grade//10*10, 90) for grade in grades)
plt.bar([x+5 for x in histogram.keys()],
       histogram.values(),
        linewidth=30,
       edgecolor='yellow',
       color='green')
plt.axis([-5,100,0,5])
plt.xticks([10*i for i in range(10)])
plt.ylabel('Grades')
plt.title('Grades vs Students')
plt.show()
```

```python
from collections import defaultdict
lists = defaultdict(list)
lists = {
    'Grades' : [5, 4, 3, 1, 4.7],
    'Scores' : [100, 70, 60, 20, 56],
    'Names'  : ['S', 'K' ,'J' ,' L', 'K']
}
```

```python
import pandas as pd
```

```python
DataFrame = pd.DataFrame(lists)
DataFrame['Class'] = [1,0,0,1,1]
DataFrame
```

```python
for label in DataFrame.columns[:]:
    plt.hist(DataFrame[DataFrame['Class']==1][label],color='red',label='Ones',alpha=0.7,density=True)
    plt.hist(DataFrame[DataFrame['Class']==0][label],color='green',alpha=0.7,label='Zeros',density=True)
    plt.xlabel(label)
    plt.ylabel("Probability")
    plt.legend()

    plt.show()
```

# Line Charts

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x+y for x,y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]  # this returns a iterator and value 
# use (i,j) for i,j in enumerate(variance) this returns a iterator and the corresponding value
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias_squared')
plt.plot(xs, total_error, 'b:', label='total_error')
```

```python
x = [1, 2 , 3, 4, 5, 66]
y = [23, 23, 3, 32, 44, 23]
z = [x+y for x,y in zip(x,y)]
count = [i for i,_ in enumerate(x)]
plt.plot(count, x, 'g-', label='x')
plt.plot(count, y, 'r-.', label='y')
plt.plot(count, z, 'b:', label='z')
plt.show()
```

# SCattere Plots

```python
friends = [ 70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [ 175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e','f','g','h','i']

for label, friend_count, minute_count in zip( labels, friends, minutes):
    plt.annotate(label,
              xy = (friend_count, minute_count),
              xytext=(5,-5),
              textcoords='offset points')
plt.scatter(friends, minutes)
plt.xlabel("Friends")
plt.ylabel("No of minutes he spent on tv ")
plt.show()
```

```python
marks = [ 12, 16, 52, 23, 98, 35]
pf_class = [ 1, 3, 4, 2, 3, 5 ]
roll_no = [1 , 2, 3, 4, 5, 6]

for roll, pf,mark in zip(roll_no, pf_class, marks):
    plt.annotate( roll,
                xy = (pf, mark),
                xytext=(5,-5),
                textcoords ='offset points'
              )
plt.scatter(pf_class,marks)
plt.title('Metrics of performance in class and Marks obtained in exams')
plt.xlabel('Performance in class')
plt.ylabel('Marks')
plt.show()
```
